```python
import nltk
import json
import codecs
%pylab inline
from os import path
from wordcloud import WordCloud
from pprint import pprint
import matplotlib.pylab as plt
```

```python
nltk.download("stopwords")
spanish_stopwords = set(nltk.corpus.stopwords.words("spanish"))
```

```python
#json_data = open("../datasets/aptitus").read()
json_data = codecs.open('../datasets/indeedd','r','utf8').read()
data = json.loads(json_data)
# pprint(data)
descripciones = [i["description"] for i in data]
text = ' '.join(descripciones)
wc = WordCloud(background_color="white", max_words=100, stopwords=map(lambda x:x.encode("utf-8","ignore"),spanish_stopwords)).generate(text)
```

```python
def showWordCloud(wcc, title, size=(25,10)):
    plt.figure(figsize=size)
    plt.title(title)
    plt.imshow(wcc)
    plt.axis("off")
    plt.show()
```

```python
showWordCloud(wc,"Aptitus")
```
