```python
import nltk
import json
import codecs
%pylab inline
from os import path
from wordcloud import WordCloud
from pprint import pprint
import matplotlib.pylab as plt
```

```python
nltk.download("stopwords")
spanish_stopwords = set(nltk.corpus.stopwords.words("spanish"))
```

```python
#json_data = open("../datasets/aptitus").read()
json_data = codecs.open('../datasets/aptitus','r','utf8').read()
data = json.loads(json_data)
# pprint(data)
descripciones = [i["description"] for i in data]
text = ' '.join(descripciones)
#udata=data.decode("utf-8")
#data=udata.encode("latin-1","ignore")
wc = WordCloud(background_color="white", max_words=100, stopwords=spanish_stopwords).generate(text)
```

```python
def showWordCloud(wcc, title, size=(15,30)):
    plt.figure(figsize=size)
    plt.title(title)
    plt.imshow(wcc)
    plt.axis("off")
    plt.show()
    
```

```python
showWordCloud(wc,"Aptitus")
```
